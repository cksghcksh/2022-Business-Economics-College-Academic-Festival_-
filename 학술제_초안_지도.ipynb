```python
pal = pal.replace('창신제1동','창신1동')
pal = pal.replace('창신제2동','창신2동')
pal = pal.replace('창신제3동','창신3동')
pal = pal.replace('숭인제1동','숭인1동')
pal = pal.replace('숭인제2동','숭인2동')
```
